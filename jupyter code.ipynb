### Data insights

# -1 Old customers ages

```python
#working on c_d / c_a / t
#children 0:14
#youth 15:24
#adults 25:64
#seniors above 65
```

```python
#c_d['DOB']=pd.to_datetime(c_d['DOB'])
c_d['age']=2021-c_d['DOB'].dt.year
c_d=c_d.dropna(axis=0)
c_d.reset_index(drop=True)
c_d=c_d.sort_values(['customer_id'],ascending=True)
c_d['age']=pd.to_numeric(c_d['age'])


children=c_d.loc[c_d['age']<14]

children_earnings=pd.merge(children, result, on=['customer_id'], how='inner')


youth=c_d.loc[(c_d['age']>14)&(c_d['age']<25)]

youth_earnings=pd.merge(youth, result, on=['customer_id'], how='inner')


adults=c_d.loc[(c_d['age']>24)&(c_d['age']<65)]

adults_earnings=pd.merge(adults, result, on=['customer_id'], how='inner')


seniors=c_d.loc[c_d['age']>64]

seniors_earnings=pd.merge(seniors, result, on=['customer_id'], how='inner')




children_earnings.to_csv('children.csv')
youth_earnings.to_csv('youth.csv')
adults_earnings.to_csv('adults.csv')
seniors_earnings.to_csv('seniors.csv')
```
